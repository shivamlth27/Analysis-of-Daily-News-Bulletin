- ### The Youtube Transcript API can get the youtube video subtitles/transcripts by giving the video id.
The
- ### The Video id of a Youtube video is the string which comes in the Browser path of video i.e. //youtube?v="__This right here__".
- ### The code given below calls the api and gets the result and then adds all the separated transcripts together to give a compbine string.
- ### The reason why those separated transcripts were not taken together because they come randomly and not according to a particular order.

```python
from youtube_transcript_api import YouTubeTranscriptApi as api

# Enter the video id in the variable given below
id='KIfuydYZXf0'
data=api.get_transcript(id, languages=['en','hi'])

print(data)
para=''
for elm in data:
  para=para+elm['text']+' '

para
```
