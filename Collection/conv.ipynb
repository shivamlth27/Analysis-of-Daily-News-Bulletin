```python
import pandas as pd
from googletrans import Translator
from datetime import datetime
id=''
name=''; date=''; nat='';
data=None
with open('temp.txt','r') as obj:

  data=[ str(elm) for elm in obj.readlines()]

  temp=data[0].split(',');
  name=name+temp[0]; date=date+temp[1]; nat=nat+temp[2]
  data.remove(data[0])

dtf=[ int(elm) for elm in date.split('-') ]
new_datetm=datetime(dtf[2],dtf[1],dtf[0])
new_datetm=new_datetm.strftime('%x')

dt={'News':data, 'Channel':[], 'Country':[],'Date':[]}
for i in range( len(data) ):
  dt['Channel'].append(name); dt['Country'].append(nat); dt['Date'].append(new_datetm)

df=pd.DataFrame(dt)
df
```

- Here, we use the translator to convert the Hindi news to Engish.

```python
translator = Translator()

for ind, row in df.iterrows():

  # Converting each news headline into a line.
  temp=  translator.translate(df.loc[ind,'News']).text
  df.loc[ind, 'News']=temp

df
```

Now we store it in a file known as collective.csv.

```python
df2=pd.read_csv('collective.csv')
df3=pd.concat([df,df2])

data=df3.to_csv('collective.csv',mode='w+',index=False)
df3
```
